## Conditional Statements (if/elif/else)

#### Age

```python
age=int(input("Enter the age: "))
if age>=18:
    print("You're Older")
else:
    print("You're Minor")
```

#### Mark

```python
Marks=int(input("Enter the marks: "))
if Marks>50:
    print("You're Pass")
else:
    print("You're Fail")
```

#### Grading System logic

```python
marks=int(input("Enter the marks of the subject: "))
if marks>=75:
    print("Grade A")
elif marks>=65:
    print("Grade B")
elif marks>=50:
    print("Grade C")
else:
    print("Grade F")
```

#### Q3 Grading System

```python
name=input("Enter your name here: ")
marks=int(input("Enter the marks of the subject: "))
grade=""

if marks>=75:
    grade = "D"
elif marks>=65:
    grade = "C"
elif marks>=50:
    grade = "B"
else:
    grade = "F"
print(name, " " , marks, " ", "You have", grade, "pass")
```

#### Example 3: Nested if Statements

```python
x=10
y=20

if x>5:
    if y>15:
        print("Both conditions are true")
    else:
        print("Only x > 5 is true")
else:
    print("x is not greater than 5")
```

#### CCT: Grade Evaluator 

```python
score = int(input("Enter your exam score (0-100): "))

if score<=100:
    if score>=90: 
        print("Grade: A")
    elif score>=80:
        print("Grade: B")
    elif score>=70:
        print("Grade: C")
    elif score>=60:
        print("Grade: D")
    elif score>=0:
        print("Grade: F")
else:
    print("Invalid score entered.")
```
